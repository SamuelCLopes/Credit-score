```python
# Importando bibliotecas

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregando base de dados

df = pd.read_csv('b1abb33c4f.csv')
df
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[df['APGAR5'] <= 3.0, 'APGAR_CAT'] = 'Asfixia Severa'
df.loc[(df['APGAR5'] >= 4.0) & (df['APGAR5'] <= 5.0), 'APGAR_CAT'] = 'Asfixia Moderada'
df.loc[(df['APGAR5'] >= 6.0) & (df['APGAR5'] <= 7.0), 'APGAR_CAT'] = 'Asfixia Leve'
df.loc[df['APGAR5'] >= 8, 'APGAR_CAT'] = 'Normal'

df
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
ax = sns.boxplot(data = df, y="APGAR5", x="SEXO", orient="v", hue='GESTACAO')
```

```python
ax1 = sns.catplot(x='GESTACAO' # definindo o eixo x
            , y='APGAR5' # definindo o eixo y
            , hue = 'SEXO'
            , kind='bar' # definindo o tipo do gráfico (barras)
            , data=df # filtrando somente as gorjetas menores que 100%
            , errorbar = None
           )

ax1.tick_params(axis="x", rotation = 30) # rotacionando eixo 'x' em 30º
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
freq = pd.crosstab(df['GESTACAO'], df['APGAR5'])
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_pctlin = freq.div(freq.sum(axis = 1), axis = 0) # Está pegando a soma do total de fumantes e dividindo pela quantidade no dia da semana.
freq_pctlin
```

```python
freq_pctlin.sum(axis = 1)
```
